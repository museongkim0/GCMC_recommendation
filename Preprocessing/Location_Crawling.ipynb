```python
import pandas as pd
import numpy as np

user_df = pd.read_csv('User_Features.csv', index_col=0)
user_df.index = range(len(user_df))
```

```python
def modify_location(loc_str):
    loc_list = loc_str.replace("n/a","").replace("/","").replace("  "," ").split(",")
    location = ""
    for loc in loc_list:
         non_space_loc = loc.replace(" ","")
         if len(non_space_loc) >= 1 and non_space_loc[0].isalpha():
              if loc != loc_list[-1]:
                    loc = loc.strip() + ","
                    location += loc
              else:
                   location += loc.strip()
    return location
```

```python
user_df['Location'] = user_df['Location'].apply(modify_location)
user_df
```

```python
import time
from selenium import webdriver

def return_lat_long(loca):
    webdriver_options = webdriver.ChromeOptions()
    webdriver_options .add_argument('headless')
    driver = webdriver.Chrome(options=webdriver_options)

    #url 생성, 검색어: location
    baseUrl = 'https://www.google.co.kr/maps/place/'
    location = loca

    #해당 url로 이동, 로딩 이후 새로운 url에 위도/경도 정보가 표시됨
    url = baseUrl + location
    driver.get(url)

    time.sleep(5)

    #현재 url에서 위도, 경도 정보 추출
    curr_url = driver.current_url
    lat_long =curr_url.split("/")[-2].split(",")

    latitude = lat_long[0][1:]
    longitude =lat_long[1]

    print("location:" , location, ", latitude: {}, longitude {}".format(latitude,longitude))
    return latitude, longitude
    driver.close()
```

```python
lat_lon_df = pd.DataFrame(columns=['Latitude','Longitude'])

try:
    for idx, loca in enumerate(user_df['Location']):
        print("index: ",idx)
        lati, longi = return_lat_long(loca)
        lat_lon_df.loc[idx] = [lati,longi]
except:
    print("error occur! save data")
    lat_lon_df.to_csv('lat_lon_df_{}'.format(idx))
```

```python
try:
    str_point = len(lat_lon_df)
    for idx, loca in enumerate(user_df['Location'][str_point:]):
        print("index: ",str_point+idx)
        lati, longi = return_lat_long(loca)
        lat_lon_df.loc[str_point + idx] = [lati,longi]
except:
    print("error occur! save data")
    lat_lon_df.to_csv('lat_lon_df_{}'.format(str_point+idx))
```
